```python
# import pandas lib as pd
import pandas as pd
# read by default 1st sheet of an excel file
from openpyxl import Workbook
import os
branch_code=7

# dataframe1 =pd.read_html('GTU/RESULT/'+'2019'+'/'+'DIPL SEM 6 - Regular.xls')
# #UNIT_NO
# df=pd.DataFrame(dataframe1[0])
# #print(df)
# comp_df=df[df['UNIT_NO'].astype(str).str.match('^16') | df['UNIT_NO'].astype(str).str.match('^17')]

# print(comp_df['CPI'].mean())
for year in ['2019']:#'2015','2016','2017','2018',
    df_list={}
    try:
        for file in os.listdir('GTU/RESULT/'+year+'/'):
            print(file.__str__())
            dataframe1 =pd.read_html('GTU/RESULT/'+year+'/'+file.__str__())
            #UNIT_NO
            df=pd.DataFrame(dataframe1[0])
            #print(df)
            comp_df=df[df['BR_CODE']==branch_code]
            comp_df=comp_df.reset_index()
            #comp_df=comp_df[comp_df['CURBACKL']==0]
            #comp_df=comp_df.reset_index()
            df_list[file.__str__().split('.')[0]]=comp_df
            if len(comp_df)==0:
                continue
            print(comp_df)
            workbook = Workbook()
            sheet = workbook.active
            #sheet.title=comp_df.loc[0,'exam']
            sheet.title=comp_df.loc[0,'sem']
            row_id=2
            column_id=1
            sheet.cell(row=1, column=1).value = "SR.NO."
            sheet.cell(row=1, column=2).value = "MAP_NUMBER"
            sheet.cell(row=1, column=3).value = "name"
            '''    if not(pd.isna(comp_df.loc[0,'SUB1'])):
                sheet.cell(row=1, column=4).value = comp_df.loc[0,'SUB1NA']+'('+comp_df.loc[0,'SUB1'].__str__()+')'
            if not(pd.isna(comp_df.loc[0, 'SUB2'])):
                sheet.cell(row=1, column=5).value = comp_df.loc[0,'SUB2NA']+'('+comp_df.loc[0,'SUB2'].__str__()+')'
            if not(pd.isna(comp_df.loc[0, 'SUB3'] )):
                sheet.cell(row=1, column=6).value = comp_df.loc[0,'SUB3NA']+'('+comp_df.loc[0,'SUB3'].__str__()+')'
            if not(pd.isna(comp_df.loc[0, 'SUB4'] )):
                sheet.cell(row=1, column=7).value = comp_df.loc[0,'SUB4NA']+'('+comp_df.loc[0,'SUB4'].__str__()+')'
            if not(pd.isna(comp_df.loc[0, 'SUB5'] )):
                sheet.cell(row=1, column=8).value = comp_df.loc[0,'SUB5NA']+'('+comp_df.loc[0,'SUB5'].__str__()+')'
            if isinstance(comp_df.loc[0, 'SUB6'],int):
                sheet.cell(row=1, column=9).value = comp_df.loc[0,'SUB6NA']+'('+comp_df.loc[0,'SUB6'].__str__()+')'
            '''
            sheet.cell(row=1, column=4).value = comp_df.loc[0, 'SUB1']
            sheet.cell(row=1, column=5).value = comp_df.loc[0, 'SUB2']
            sheet.cell(row=1, column=6).value = comp_df.loc[0, 'SUB3']
            sheet.cell(row=1, column=7).value = comp_df.loc[0, 'SUB4']
            sheet.cell(row=1, column=8).value = comp_df.loc[0, 'SUB5']
            sheet.cell(row=1, column=9).value = comp_df.loc[0, 'SUB6']

            sheet.cell(row=1, column=10).value ='TOTBACKL'
            sheet.cell(row=1, column=11).value ='CURBACKL'
            sheet.cell(row=1, column=12).value ='SPI'
            sheet.cell(row=1, column=13).value ='CPI'
            sheet.cell(row=1, column=14).value ='CGPA'
            sheet.cell(row=1, column=15).value ='RESULT'

            count=1
            for i in range(0, len(comp_df)):
                sheet.cell(row=row_id, column=1).value = row_id-1
                sheet.cell(row=row_id, column=2).value = comp_df.loc[i,'MAP_NUMBER']
                sheet.cell(row=row_id, column=3).value = comp_df.loc[i,'name']
                sheet.cell(row=row_id, column=4).value = comp_df.loc[i,'SUB1GR']
                sheet.cell(row=row_id, column=5).value = comp_df.loc[i,'SUB2GR']
                sheet.cell(row=row_id, column=6).value = comp_df.loc[i,'SUB3GR']
                sheet.cell(row=row_id, column=7).value = comp_df.loc[i,'SUB4GR']
                sheet.cell(row=row_id, column=8).value = comp_df.loc[i,'SUB5GR']
                sheet.cell(row=row_id, column=9).value = comp_df.loc[i,'SUB6GR']
                sheet.cell(row=row_id, column=10).value = comp_df.loc[i,'TOTBACKL']
                sheet.cell(row=row_id, column=11).value = comp_df.loc[i,'CURBACKL']
                sheet.cell(row=row_id, column=12).value = comp_df.loc[i,'SPI']
                sheet.cell(row=row_id, column=13).value = comp_df.loc[i,'CPI']
                sheet.cell(row=row_id, column=14).value = comp_df.loc[i,'CGPA']
                sheet.cell(row=row_id, column=15).value = comp_df.loc[i,'RESULT']

                row_id=row_id+1
                workbook.save(filename=('GTU/RESULT_ANALYSIS/'+year+'/'+file.__str__()))

    except:
        print('exec()')
```
